```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/FRANK/Desktop/final_sample.csv")
```

```python
df.close_price.plot()
```

```python
df_features = df.filter(like='feature')
df_features.columns.shape
```

```python
df_categorical = df_features.filter(like='integer')
df_continous = df_features[df_features.columns.difference(df_categorical.columns)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up  matplotlib figure
f, ax = plt.subplots(figsize=(25, 20))

# calculate the correlation matrix
corr = df_continous.corr()

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,ax=ax)
```

```python
for feature in df_continous.columns:
    print(f"   Plotting feature: '{feature}'")
    tempSet = df_continous[feature]
    print(f"   Total data length... {len(tempSet)}")
    plt.plot(tempSet)
    plt.show()
    print("***"*15)
```
